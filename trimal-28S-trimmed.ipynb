## Plot Positional Residue Similarity Score

```python
%pylab inline
```

```python
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt

# read in data, skip first 3 rows as this is header
# some colums are separated by two tabs hence \t+ as seperator
var = pd.read_table('28S_trimmed_aln_ssc.txt', sep='\t+', skiprows=3)
var.columns = ['position', 'variability']

#Plot residue similarity figure, for nucleotides this is identity value 
figure(figsize=(24,4), dpi=80)
var.variability.plot(color='g',lw=2)
title("Trimal's Residue Similarity Score (-ssc)\n ", fontsize=20)
xlabel("alignment position", fontsize=16)
ylabel("similarity score", fontsize=16)
axis([1,711, 0, 1.1]) #0 to 1 scale for y axis
grid(True)
show()
```
