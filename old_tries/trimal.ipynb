# Parse similarity and gap data from Trimal

## Run Trimal on fasta alignment file

```python
!trimal -in ./data/22rots.afa -ssc > 22rots_vari_ssc.txt
```

```python
!head 22rots_vari_ssc.txt
```

## Plot Positional Residue Similarity Score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read in data, skip first 3 rows as this is header
# some colums are separated by two tabs hence \t+ as seperator
var = pd.read_table('22rots_vari_ssc.txt', sep='\t+', skiprows=3)
var.columns = ['position', 'variability']

#Plot residue similarity figure, for nucleotides this is identity value 
figure(figsize=(24,4), dpi=80)
var.variability.plot(color='g',lw=2)
plt.title("Trimal's Residue Similarity Score (-ssc)\n ", fontsize=20)
plt.xlabel("alignment position", fontsize=16)
plt.ylabel("similarity score", fontsize=16)
plt.axis([1,711, 0, 1.1]) #0 to 1 scale for y axis
plt.grid(True)
plt.show()
```

## Gap frequency plot

```python
!trimal -in ./data/22rots.afa -sgc > 22rots_vari_sgc.txt
```

```python
!head 22rots_vari_sgc.txt
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read in data, skip first 3 rows as this is header
# some colums are separated by two tabs hence \t+ as seperator
var = pd.read_table('22rots_vari_sgc.txt', sep='\t+', skiprows=3)
var.columns = ['position', 'pct_gaps', 'gap_score']

#print(var[:10]) #print first 10 lines of input file

#Plot percent gaps figure
figure(figsize=(24,4), dpi=80)
var.pct_gaps.plot(color='g',lw=2)
plt.title("Trimal's column percent gaps (-sgc)", fontsize=20)
plt.xlabel("alignment position", fontsize=16)
plt.ylabel("percent gaps", fontsize=16)
plt.axis([1, 711, 0, 110]) #0 to 100 scale for y axis, ie percent
plt.grid(True)
plt.show()

#Plot gap score figure
figure(figsize=(24,4), dpi=80)
var.gap_score.plot(color='r',lw=2)
plt.title("Trimal's column gap gcore (-sgc)", fontsize=20)
plt.xlabel("alignment position", fontsize=16)
plt.ylabel("gap score\n(fraction no gaps)", fontsize=16)
plt.axis([1, 711, 0, 1.1]) #0 to 1 scale for y axis
plt.grid(True)
plt.show()
#print('Gap score: The gap score for a column is the fraction of positions in the column without a gap. Higher values are less gappy columns.')
```
