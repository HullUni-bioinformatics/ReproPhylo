```python
from Bio import SeqIO
sizes = [len(rec) for rec in SeqIO.parse("data/22_rotifer_COI_seqs.fas", "fasta")]
import pylab
pylab.hist(sizes, bins=40)
pylab.title("%i Brachionus sequences\nLengths %i to %i" \
	% (len(sizes),min(sizes),max(sizes))) 
pylab.xlabel("Sequence length (bp)")
pylab.ylabel("Count") 
pylab.show()
```

```python
from Bio import SeqIO
from Bio.SeqUtils import GC

gc_values = sorted(GC(rec.seq) for rec in SeqIO.parse("data/22_rotifer_COI_seqs.fas", "fasta"))

import pylab
pylab.plot(gc_values)
pylab.title("%i brachionus\nGC%% %0.1f to %0.1f" \
            % (len(gc_values),min(gc_values),max(gc_values)))
pylab.xlabel("Genes")
pylab.ylabel("GC%")
pylab.show()
```

```python
#this modified from BioPython tutorial
from Bio import SeqIO 
sizes = [len(rec) for rec in SeqIO.parse("data/22_rotifer_COI_seqs.fas", "fasta")]
import pylab
#not specifying color also gives blue
pylab.hist(sizes, bins=40, color='blue')
pylab.title("Lengths of %i Brachionus rotifer sequences\n Mean %i bp, range %i - %i bp" \
	% (len(sizes),mean(sizes),min(sizes),max(sizes))) 
pylab.xlabel("Sequence length (bp)")
pylab.ylabel("Count") 
pylab.show()

#this from BioPython tutorial
from Bio import SeqIO
from Bio.SeqUtils import GC

gc_values = sorted(GC(rec.seq) for rec in SeqIO.parse("data/22_rotifer_COI_seqs.fas", "fasta"))

#this put together by me
pylab.hist(gc_values, bins=40, color='green')
pylab.title("Percentage GC for %i Brachionus rotifer sequences\n Mean %0.1f%%, range %0.1f - %0.1f%%" \
	% (len(gc_values),mean(gc_values),min(gc_values),max(gc_values)))
pylab.xlabel("Percentage GC")
pylab.ylabel("Count") 
pylab.show()
```
