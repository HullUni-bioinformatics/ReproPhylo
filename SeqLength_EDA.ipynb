Here is everything together, producing two graphs

```python
#this modified from BioPython tutorial
from Bio import SeqIO 
sizes = [len(rec) for rec in SeqIO.parse("data/brachionus2k.fasta", "fasta")]

#not specifying color also gives blue
hist(sizes, bins=40, color='blue')
title("Lengths of %i Brachionus rotifer sequences\n Mean %i bp, range %i - %i bp" % (len(sizes), mean(sizes), min(sizes), max(sizes))) 
xlabel("Sequence length (bp)")
ylabel("Count") 
show()

#this from BioPython tutorial
from Bio import SeqIO
from Bio.SeqUtils import GC

gc_values = sorted(GC(rec.seq) for rec in SeqIO.parse("data/brachionus2k.fasta", "fasta"))

#this put together by me
hist(gc_values, bins=40, color='green')
title("Percentage GC for %i Brachionus rotifer sequences\n Mean %0.1f%%, range %0.1f - %0.1f%%" % (len(gc_values), mean(gc_values), min(gc_values), max(gc_values)))
xlabel("Percentage GC")
ylabel("Count") 
show()
```

Try to extract range of sequences, ie ask user to enter size range to exclude/include or GC range to keep

```python
# Extract a subset of sequences as follows

# Set the range of GC values to extract
gc_subset = [val for val in gc_values if val >= 35 and val <= 40]

# Plot a histogram of extracted %GC range
hist(gc_subset, bins=40, color='green')
title("Percentage GC for %i Brachionus rotifer sequences\n Mean %0.1f%%, range %0.1f - %0.1f%%" % (len(gc_values), mean(gc_values), min(gc_values), max(gc_values)))
xlabel("Percentage GC")
ylabel("Count") 
show()

# Set the range length values to extract
len_subset = [val for val in sizes if val >= 500 and val <= 900]
#print len_subset

# Plot a histogram of extracted length range
hist(len_subset, bins=40, color='blue')
title("Length for %i Brachionus rotifer sequences\n Mean %0.1f%%, range %0.1f - %0.1f%%" % (len(gc_values), mean(gc_values), min(gc_values), max(gc_values)))
xlabel("Length")
ylabel("Count") 
show()
```
