```python
%%file tree_annotation.py

#######################################################################
#                        COMMAND LINE PARSING
#######################################################################
import argparse as agp

desc = ("Annotate all the trees in the provided Project")
parser = agp.ArgumentParser(description=desc)

#----------------------------------------------------------------------
parser.add_argument("projfile",
                    help="A pickle file to read and write",
                    type=str,
                    default=None)
parser.add_argument("figs",
                    help="A directory to keep fig files",
                    type=str, 
                    default='.')
parser.add_argument("-a", "--altprojfile",
                    help="An alternative pickle file to write",
                    type=str, 
                    default=None)
parser.add_argument("-_", "--__",
                    help=("Turn off warnings for Galaxy"),
                    action="store_true")
parser.add_argument("--noninteractive",
                    help=("Overwrite reports without asking"),
                    action="store_true")
#----------------------------------------------------------------------
parser.add_argument("-m", "--rootmeta",
                    help="Metadata by which to dermine the outgroup",
                    type=str, 
                    default='mid')
parser.add_argument("-r", "--rootvalue",
                    help=("Within the metadata column, determine a value"+
                          " that specifies the outgroup"),
                    type=str, 
                    default='mid')
parser.add_argument("-l", "--leaflabelmeta",
                    help=("Metadata to use as labels"),
                    type=str, nargs='+',
                    default=['source_organism'])
parser.add_argument("-t", "--leafcolormeta",
                    help=("Metadata to use to determine label colours"),
                    type=str, 
                    default=None)
parser.add_argument("-s", "--leafcolors",
                    help=("Pairs of metadata value and it's color"),
                    type=str, nargs='+',
                    default=None)
parser.add_argument("-g", "--nodebgmeta",
                    help=("Metadata to determine background colors"),
                    type=str,
                    default=None)
parser.add_argument("-i", "--nodebgcolors",
                    help=("Pairs of metadata value and it's color"),
                    type=str, nargs='+',
                    default=None)
parser.add_argument("-n", "--nodesupport",
                    help=("Triplets of values: eg, black 100 90 gray 90 80"),
                    type=str, nargs='+',
                    default=['black', '100', '99',
                             'dimgray', '99', '75',
                             'silver', '75', '50'])
parser.add_argument("-b", "--heatmapmeta",
                    help=("A list of metadata column names"),
                    type=str, nargs='+',
                    default=None)
parser.add_argument("-d", "--heatmapcolor",
                    help=("Heat map color scheme (0,1 or 2)"),
                    type=str, choices=['0','1','2'], 
                    default="2")
parser.add_argument("-u", "--multifurc",
                    help=("Support value under which to collapse node"),
                    type=str, 
                    default=None)
parser.add_argument("-e", "--treescale",
                    help=("Tree width"),
                    type=str, 
                    default="1000")
#----------------------------------------------------------------------
report = parser.add_mutually_exclusive_group()
report.add_argument("-v", "--verbose_report",
                   help=("Write full html report to filename"),
                   type=str,
                   default=None)
report.add_argument("--html",
                   help=("Write html of fig file links"),
                   type=str,
                   default=None)
#----------------------------------------------------------------------
parser.add_argument("-c", "--version_ctrl",
                    help=("perform version control in CWD"),
                    action="store_true")
#----------------------------------------------------------------------
cl = parser.parse_args()
#----------------------------------------------------------------------
# Coloring labels requires both metadata column and colors
if len([x for x in (cl.leafcolormeta,
                    cl.leafcolors) if x is not None])== 1:
    parser.error('To color labels, both --leafcolormeta and'+
                 ' --leafcolors are required')
#----------------------------------------------------------------------
# Coloring backgrounds requires both metadata column and colors
if len([x for x in (cl.nodebgmeta,
                    cl.nodebgcolors) if x is not None])== 1:
    parser.error('To color backgrounds, both --nodebgmeta and'+
                 ' --nodebgcolors are required')
#----------------------------------------------------------------------
if not len(cl.nodesupport)%3==0:
    parser.error('Node support annotation requires triplets consisting of: '+
                 'color top_range bottom_range')
if cl.leafcolors and not len(cl.leafcolors)%2==0:
    parser.error('--leafcolors requires couples consisting of: '+
                 'matadata_value color') 
if cl.nodebgcolors and not len(cl.nodebgcolors)%2==0:
    parser.error('--nodebgcolors requires couples consisting of: '+
                 'matadata_value color')
#----------------------------------------------------------------------
from reprophylo import *

def chekpoint_project(pj):
    if cl.altprojfile:
        pickle_pj(pj, cl.altprojfile)
    else:
        pickle_pj(pj, cl.projfile)


def dict_from_list_pairs(t):
    if t:
        dictionary = {}
        pairs = zip(t[::2], t[1::2])
        for i in pairs:
            dictionary[i[0]] = i[1]
        return dictionary
    else:
        return None
    
def dict_from_list_triplets(t):
    if t:
        dictionary = {}
        trios = zip(t[::3], t[1::3], t[2::3])
        for i in trios:
            dictionary[i[0]] = [int(i[1]),int(i[2])]
        return dictionary
    else:
        return None
    
def int_or_none(val):
    try:
        return int(val)
    except:
        return None

def removetype(path):
    import shutil
    if os.path.isdir(path):
        shutil.rmtree(path)
    else:
        os.remove(path)

#######################################################################
#                       Execution
#######################################################################
cline = ''

for i in sys.argv:
    cline += i+' '
print 'ReproPhylo was called with:'
print cline
print

if cl.__:
    warnings.simplefilter('ignore')

if cl.version_ctrl:
    start_git()
    
pj = unpickle_pj(cl.projfile)
pj.clear_tree_annotations()
pj.annotate(cl.figs,
            cl.rootmeta,
            cl.rootvalue,
            cl.leaflabelmeta,
            leaf_node_color_meta=cl.leafcolormeta,
            leaf_label_colors=dict_from_list_pairs(cl.leafcolors),
            node_bg_meta=cl.nodebgmeta,
            node_bg_color=dict_from_list_pairs(cl.nodebgcolors),
            node_support_dict=dict_from_list_triplets(cl.nodesupport),
            heat_map_meta=cl.heatmapmeta,
            heat_map_colour_scheme=cl.heatmapcolor,
            multifurc=int_or_none(cl.multifurc),
            scale=int(cl.treescale),
            html=cl.html
            )

if cl.verbose_report:
    if cl.noninteractive:
        if os.path.exists(cl.verbose_report):
            removetype(cl.verbose_report)
        if os.path.exists(cl.verbose_report.rpartition('.')[0]):
            removetype(cl.verbose_report.rpartition('.')[0])
        if os.path.exists(cl.verbose_report+'.zip'):
            removetype(cl.verbose_report+'.zip')
    publish(pj, cl.verbose_report, cl.figs)


chekpoint_project(pj)
```

```python
%%file annotate_trees.xml

<tool id="Annotate Trees" name="Annotate Trees"  version="0.1" hidden="false">
    <description>
    </description>
    <command interpreter="python">
        tree_annotation.py 
        $projin $figsdir  -_ -a $altprojout 
        #if str($root.howtoroot)=="other"
        --rootmeta $root.rootqual --rootvalue $root.rootvalue
        #end if
        #if str($labeltext.howtolabel)=="other"
        --leaflabelmeta $labeltext.leaflabelmeta
        #end if        
        #if str($labelcolor.howtocolorlabel)=="other"
        --leafcolormeta $labelcolor.leafcolormeta --leafcolors $labelcolor.leafcolors
        #end if
        #if str($bgcolor.howtocolorbg)=="yes"
        --nodebgmeta $bgcolor.nodebgmeta --nodebgcolors $bgcolor.nodebgcolors
        #end if
        #if str($nodeSupport.howtoshowsupport)=="no"
        --nodesupport $nodeSupport.nodesupport
        #end if 
        #if str($heatmap.doheatmap)=="yes"
        --heatmapmeta $heatmap.heatmapmeta --heatmapcolor $heatmap.heatmapcolor
        #end if  
        #if str($multifurc.collapse)=="yes"
        --multifurc $multifurc.multifurccutoff 
        #end if 
        --treescale $treescale 
        #if (str($whatreport)=="full")
        -v report_arcive 
        #end if
        #if (str($whatreport)=="html")
        --html figs.html 
        #end if
        > $log
    </command>
    <inputs>
        <param name="projin" type="data" label="Input Project file" />
        <param name="figsdir" type="text" label="Local path to keep figures" />
        <param name="treescale" type="text" value="1000" label="Tree width" />
        <conditional name="root">
            <param name="howtoroot" type="select" label="How to root the tree?" >
                <option value="mid">Root at midpoint</option>
                <option value="other">Specify outgroup using metadata</option>
            </param>
            <when value="mid"/>
            <when value="other">
                <param name="rootqual" type="text" label="A metadata column name" />
                <param name="rootvalue" type="text" label="The value to look for in this column" />
            </when>
        </conditional>
        <conditional name="labeltext">
            <param name="howtolabel" type="select" label="How to label leaves?" >
                <option value="orng">Label leaves with the species name</option>
                <option value="other">Specify labels using metadata</option>
            </param>
            <when value="orng"/>
            <when value="other">
                <param name="leaflabelmeta" type="text" label="A space delimited list of one or more metadata column name to use for labels" />
            </when>
        </conditional>
        <conditional name="labelcolor">
            <param name="howtocolorlabel" type="select" label="How to color labels?" >
                <option value="black">Black</option>
                <option value="other">According to rules</option>
            </param>
            <when value="orng"/>
            <when value="other">
                <param name="leafcolormeta" type="text" label="A metadata column name to use for determining color" />
                <param name="leafcolors" type="text" label="A space delimited list of one or more pairs of value and color" />
            </when>
        </conditional>
        <conditional name="bgcolor">
            <param name="howtocolorbg" type="select" label="Color node backgrounds?" >
                <option value="no">No</option>
                <option value="yes">Yes</option>
            </param>
            <when value="no"/>
            <when value="yes">
                <param name="nodebgmeta" type="text" label="A metadata column name to use for determining bg color" />
                <param name="nodebgcolors" type="text" label="A space delimited list of one or more pairs of value and color" />
            </when>
        </conditional>
        <conditional name="nodeSupport">
            <param name="howtoshowsupport" type="select" label="Use default ranges for node support annotations?" >
                <option value="yes">Yes</option>
                <option value="no">No</option>
            </param>
            <when value="yes"/>
            <when value="no">
                <param name="nodesupport" type="text" label="A space delimited list of one or more trios of color, uppper and lower limit" />
            </when>
        </conditional>
        <conditional name="heatmap">
            <param name="doheatmap" type="select" label="make a heatmap?" >
                <option value="no">No</option>
                <option value="yes">Yes</option>
            </param>
            <when value="no"/>
            <when value="yes">
                <param name="heatmapmeta" type="text" label="A list of metadata column names comprising the data vector" />
                <param name="heatmapcolor" type="select" label="Select a color scheme" >
                    <option value='0'>Green to blue</option>
                    <option value='1'>Green to red</option>
                    <option value='2'>Red to blue</option>
                </param>
            </when>
        </conditional>
        <conditional name="multifurc">
            <param name="collapse" type="select" label="Use default ranges for node support annotations?" >
                <option value="no">No</option>
                <option value="yes">Yes</option>
            </param>
            <when value="no"/>
            <when value="yes">
                <param name="multifurccutoff" type="text" label="Node support cutoff for node collapsing" />
            </when>
        </conditional>
        <param name="whatreport" type="select" label="Choose report type" >
            <option value="html">HTML with links to figures</option>
            <option value="full">Full report archive</option>
        </param>          
    </inputs>
    <outputs>
        <data name="log" format='txt' label="${tool.name} on ${on_string}: Log" />
        <data name="altprojout" format='data' label="${tool.name} on ${on_string}: ReproPhylo Project" />
        <data name="reportName" from_work_dir="report_arcive.zip" format="data" label="${tool.name} on ${on_string}: Report archive">
            <filter>whatreport=="full"</filter>
        </data>
        <data name="htmlName" from_work_dir="figs.html" format="html" label="${tool.name} on ${on_string}: Figures html">
            <filter>whatreport=="html"</filter>
        </data>
    </outputs>
    <help>
    
ReproPhylo
==========

"Annotate Trees"
----------------

</help>
</tool>    
```
